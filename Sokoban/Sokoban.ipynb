## Run Function

```python
def dqn(n_episodes=1000, max_steps=300, eps_start=1.0, eps_end=0.01, eps_decay=0.995):
    scores = []                        
    scores_window = deque(maxlen=100)  
    eps = eps_start                    

    for i_episode in range(1, n_episodes+1):
        state = np.array(ImageOps.grayscale(Image.fromarray(env.reset())))
        score = 0

        for i_step in range(max_steps):
            action = agent.act(state, eps)
            next_state, reward, done, _ = env.step(action)
            next_state = np.array(ImageOps.grayscale(Image.fromarray(next_state)))
            agent.step(state, action, reward, next_state, done)
            state = next_state
            score += reward

            if done:
                break 

        scores_window.append(score)       
        scores.append(score)              
        eps = max(eps_end, eps_decay*eps) 
        print('Episode {}\tAverage Score: {:.2f}\n'.format(i_episode, np.mean(scores_window)), end="")

        if i_episode % 100 == 0:
            print('Episode {}\tAverage Score: {:.2f}\n'.format(i_episode, np.mean(scores_window)))

        if np.mean(scores_window)>=10.0:
            print('\nEnvironment solved in {:d} episodes!\tAverage Score: {:.2f}'.format(i_episode-100, np.mean(scores_window)))
            torch.save(agent.qnetwork_local.state_dict(), 'checkpoint.pth')
            break

    return scores
```

```python
agent = Agent(state_size, action_size, 32)
scores = dqn()
```

```python
# plot the scores
fig = plt.figure(my_rewards=True)
ax = fig.add_subplot(111)
plt.plot(np.arange(len(scores)), scores)
plt.ylabel('Score')
plt.xlabel('Episode #')
plt.show()
```
